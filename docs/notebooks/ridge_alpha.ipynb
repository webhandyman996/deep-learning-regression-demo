# 嶺回歸：調整正則化強度 Alpha

拖曳 Alpha 滑桿，觀察正則化強度如何影響預測值與 MSE。

（本 notebook 已將原本本機版的 matplotlib Slider 改寫為 ipywidgets，在 Colab 網頁環境才能正常拖曳互動）

點選上方「執行階段」→「全部執行」，再拖曳滑桿互動。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from ipywidgets import interact, FloatSlider

np.random.seed(42)
n_samples = 100
ad_spend = np.random.uniform(1, 100, (n_samples, 3))
sales = 0.5 * ad_spend[:, 0] + 0.3 * ad_spend[:, 1] + 0.2 * ad_spend[:, 2] + np.random.randn(n_samples) * 10

X_train, X_test, y_train, y_test = train_test_split(ad_spend, sales, test_size=0.2, random_state=42)

def plot(alpha=3.0):
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(X_train, y_train)
    y_pred = ridge_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)

    plt.figure()
    plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Ideal Fit')
    plt.xlabel('Actual Sales (thousands)')
    plt.ylabel('Predicted Sales (thousands)')
    plt.legend()
    plt.title(f'Ridge Regression: Alpha={alpha:.1f}, MSE={mse:.2f}')
    plt.show()

interact(plot, alpha=FloatSlider(min=0.1, max=10.0, step=0.1, value=3.0, description='Alpha'))
```
