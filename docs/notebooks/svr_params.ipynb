# 支持向量回歸：調整 C / Gamma / Epsilon

拖曳三個參數滑桿，觀察 SVR 核函數參數如何影響預測結果與 MSE。

（本 notebook 已將原本本機版的 matplotlib Slider 改寫為 ipywidgets，在 Colab 網頁環境才能正常拖曳互動）

點選上方「執行階段」→「全部執行」，再拖曳滑桿互動。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from ipywidgets import interact, FloatSlider

np.random.seed(42)
n_samples = 100
features = np.random.uniform(1000, 3000, (n_samples, 3))  # 車重(kg)、引擎大小(L)、氣缸數
fuel_consumption = 0.005 * features[:, 0] + 0.8 * features[:, 1] + 0.3 * features[:, 2] + np.random.randn(n_samples) * 2

X_train, X_test, y_train, y_test = train_test_split(features, fuel_consumption, test_size=0.2, random_state=42)

def plot(C=100.0, gamma=0.1, epsilon=0.1):
    svr_model = SVR(kernel='rbf', C=C, gamma=gamma, epsilon=epsilon)
    svr_model.fit(X_train, y_train)
    y_pred = svr_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)

    plt.figure()
    plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Ideal Fit')
    plt.xlabel('Actual Fuel Consumption (L/100km)')
    plt.ylabel('Predicted Fuel Consumption (L/100km)')
    plt.legend()
    plt.title(f'SVR: C={C:.1f}, Gamma={gamma:.2f}, Epsilon={epsilon:.2f}, MSE={mse:.2f}')
    plt.show()

interact(
    plot,
    C=FloatSlider(min=1, max=200, step=1, value=100, description='C'),
    gamma=FloatSlider(min=0.01, max=1.0, step=0.01, value=0.1, description='Gamma'),
    epsilon=FloatSlider(min=0.01, max=1.0, step=0.01, value=0.1, description='Epsilon'),
)
```
