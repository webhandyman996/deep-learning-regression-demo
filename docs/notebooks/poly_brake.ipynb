# 多項式回歸：煞車距離係數調整

拖曳 a、b、c 三個係數滑桿，即時觀察煞車距離曲線與資料分佈如何變化。

### 各參數的意義與影響

- **a（x² 項係數）**：值越大，車速增加時煞車距離會以更快的加速度上升，代表高速時煞車距離對速度更敏感。
- **b（x 項係數）**：控制速度與煞車距離的線性比例，值越大代表低速時煞車距離也拉長越多。
- **c（常數項）**：即使車速趨近 0，仍會存在的基礎距離（例如反應距離）。

點選上方「執行階段」→「全部執行」，再拖曳下方滑桿互動。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from ipywidgets import interact, FloatSlider

# 模擬數據：汽車速度 (km/h) 與煞車距離 (m)
np.random.seed(42)
speed = np.random.uniform(20, 120, 50).reshape(-1, 1)

def calculate_braking_distance(a, b, c):
    return a * speed**2 + b * speed + c + np.random.randn(50, 1) * 10

def plot(a=0.05, b=2.0, c=5.0):
    braking_distance = calculate_braking_distance(a, b, c)

    poly = PolynomialFeatures(degree=2)
    speed_poly = poly.fit_transform(speed)
    model = LinearRegression()
    model.fit(speed_poly, braking_distance)

    speed_fit = np.linspace(20, 120, 100).reshape(-1, 1)
    braking_distance_fit = model.predict(poly.transform(speed_fit))

    plt.figure()
    plt.scatter(speed, braking_distance, color='blue', label='Observed Data')
    plt.plot(speed_fit, braking_distance_fit, color='red', label='Polynomial Regression Fit')
    plt.xlabel('Speed (km/h)')
    plt.ylabel('Braking Distance (m)')
    plt.legend()
    plt.title(f'Polynomial Regression: a={a:.2f}, b={b:.2f}, c={c:.2f}')
    plt.show()

interact(
    plot,
    a=FloatSlider(min=0.01, max=0.1, step=0.01, value=0.05, description='a (x^2)'),
    b=FloatSlider(min=0, max=5, step=0.1, value=2.0, description='b (x)'),
    c=FloatSlider(min=0, max=10, step=0.1, value=5.0, description='c (常數)'),
)
```
