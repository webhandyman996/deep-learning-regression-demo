# 多項式回歸：調整項次 (Degree)

拖曳 Degree 滑桿，觀察多項式次數如何影響擬合曲線的彎曲程度與方程式。

### 各參數的意義與影響

- **Degree（多項式次方）**：次方越高，曲線能彎出越複雜的形狀、更貼近訓練資料點；但次方太高時曲線會過度追逐雜訊，造成過擬合，在新資料上反而表現變差。

點選上方「執行階段」→「全部執行」，再拖曳下方滑桿互動。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from ipywidgets import interact, IntSlider

np.random.seed(42)
speed = np.random.uniform(20, 120, 50).reshape(-1, 1)
braking_distance = 0.05 * speed**2 + 2 * speed + 5 + np.random.randn(50, 1) * 10

def plot(degree=2):
    poly = PolynomialFeatures(degree=degree)
    speed_poly = poly.fit_transform(speed)
    model = LinearRegression()
    model.fit(speed_poly, braking_distance)

    speed_fit = np.linspace(20, 120, 100).reshape(-1, 1)
    braking_distance_fit = model.predict(poly.transform(speed_fit))

    coeffs = model.coef_.flatten()
    intercept = model.intercept_.flatten()[0]
    equation = f"y = {intercept:.2f}"
    for i, coef in enumerate(coeffs[1:], start=1):
        equation += f" + {coef:.2f}x^{i}"

    plt.figure()
    plt.scatter(speed, braking_distance, color='blue', label='Observed Data')
    plt.plot(speed_fit, braking_distance_fit, color='red', label='Polynomial Regression Fit')
    plt.xlabel('Speed (km/h)')
    plt.ylabel('Braking Distance (m)')
    plt.legend()
    plt.title(f'Polynomial Regression: Degree={degree}')
    plt.gca().text(0.05, 0.95, equation, transform=plt.gca().transAxes, fontsize=9, va='top')
    plt.show()

interact(plot, degree=IntSlider(min=1, max=10, step=1, value=2, description='Degree'))
```
