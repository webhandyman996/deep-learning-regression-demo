# 邏輯回歸：調整分類門檻 Threshold

拖曳 Threshold 滑桿，觀察分類門檻如何影響購買與否的判定結果。

### 各參數的意義與影響

- **Threshold（分類門檻）**：模型輸出的是「購買機率」，Threshold 是判斷「算不算購買」的分界線。值越低，越容易判定為會購買（可能增加誤判）；值越高則判斷更嚴格（可能漏掉真的會購買的顧客）。

點選上方「執行階段」→「全部執行」，再拖曳下方滑桿互動。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from ipywidgets import interact, FloatSlider

np.random.seed(42)
browse_time = np.random.uniform(1, 20, 100).reshape(-1, 1)
purchase = (browse_time + np.random.randn(100, 1) * 2 > 10).astype(int).ravel()

X_train, X_test, y_train, y_test = train_test_split(browse_time, purchase, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

browse_time_fit = np.linspace(1, 20, 100).reshape(-1, 1)
purchase_prob = model.predict_proba(browse_time_fit)[:, 1]

def plot(threshold=0.5):
    predicted_purchase = (purchase_prob >= threshold).astype(int)

    plt.figure()
    plt.scatter(browse_time, purchase, color='blue', label='Observed Data')
    plt.plot(browse_time_fit, purchase_prob, color='red', label='Logistic Regression Fit')
    plt.plot(browse_time_fit, [threshold] * len(browse_time_fit), color='green', linestyle='--', label='Threshold')
    plt.xlabel('Browse Time (minutes)')
    plt.ylabel('Purchase Probability')
    plt.legend()
    plt.title(f'Logistic Regression: Threshold={threshold:.2f}')
    plt.show()

interact(plot, threshold=FloatSlider(min=0.1, max=0.9, step=0.01, value=0.5, description='Threshold'))
```
